```python
from scipy import stats
F_statistic, pVal = stats.f_oneway(jcg_midx_jal, jcg_midx_mid, jcg_midx_mot)
print("F통계량: ",F_statistic,"유의확률:", pVal)
#print('데이터의 일원분산분석 결과 : F={0:.1f}, p={1:.10f}'.format(F_statistic, pVal))
#print('데이터의 일원분산분석 결과 : F=%.2f , p=%.10f ' %(F_statistic, pVal))
```
