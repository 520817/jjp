```python
ssang = pd.read_csv('/content/AHP용.csv')
ssang
```

```python
ssang = ssang.drop(['Unnamed: 0'],axis=1)
ssang
```

```python
# 데이터프레임의 컬럼수 파악
colums_cnt = len(ssang.columns)
colums_cnt
```

```python
# 각 컬럼의 기하평균을 계산하여 리스트에 저장
ssang_mean_list = []
for i in ssang:
    ssang_mean_list.append(gmean(ssang[i]))
```

```python
# 기하평균 출력
print('기하평균 (Geometric mean)\n')
for i,j in zip(ssang, ssang_mean_list):
    print('{} : {}'.format(i, j))
    #print('{} : {}'.format(i, round(j,3)))  # 소수점 3 자리수
```

```python
# 데이터프레임에 기하평균값 추가
ss = ssang.append(pd.Series(ssang_mean_list, index=ssang.columns),ignore_index=True)
ss
```

```python
# 가중치 계산을 위하여 기하평균의 합을 구함
ssang_mean_sum = sum(ssang_mean_list)
ssang_mean_sum
```

```python
# 각 컬럼의 가중치를 계산하여 리스트에 저장
weight_list = []
for i in range(colums_cnt):
    weight_list.append(ssang_mean_list[i]/ssang_mean_sum)
```

```python
# 기하평균 출력
print('가중치 (Weights)\n')
for i,j in zip(ss, weight_list):
    print('{} : {}'.format(i, j))
    #print('{} : {}'.format(i, round(j,3)))  # 소수점 3 자리수
```

```python
# 가중치 합 확인
su_ = sum(list(weight_list))
print(su_)
```

```python
ss = ss.append(pd.Series(weight_list, index=ss.columns),ignore_index=True)
ss
```

```python
# 데이터프레임을 csv로 저장
ss.to_csv('AHP_가중치산정_결과.csv',index=True , sep=',', encoding='cp949')
```
