```python
import pandas as pd
```

```python
survey = pd.read_csv('/content/AHP용.csv')
survey.head(5)
```

```python
survey = survey.drop(['Unnamed: 0'],axis=1)
survey.head(5)
```

```python
import numpy as np
from scipy.stats.mstats import gmean
```

```python
survey = survey.astype('float')
```

```python
# 각 컬럼의 기하평균을 계산하여 리스트에 저장
geo_mean_list = []
for i in survey:
    geo_mean_list.append(gmean(survey[i]))
```

```python
	#가중치
#기초생활수급자인원	0.302868
#노년부양비	0.178305
#고령내국인전체	0.112779
#노령화지수	0.191392
#무료급식소개수	0.214656
```

```python
# 기하평균 출력
print('기하평균 (Geometric mean)\n')
for i,j in zip(survey, geo_mean_list):
    #print('{} : {}'.format(i, j))
    print('{} : {}'.format(i, round(j,3)))  # 소수점 3 자리수
```

```python
# 데이터프레임에 기하평균값 추가
survey = survey.append(pd.Series(geo_mean_list, index=survey.columns),ignore_index=True)
survey
```

```python
# 데이터프레임의 컬럼수 파악
colums_cnt = len(survey.columns)
colums_cnt
```

```python
geo_mean_sum = sum(geo_mean_list)
geo_mean_sum
```

```python
# 각 컬럼의 가중치를 계산하여 리스트에 저장
weight_list = []
for i in range(colums_cnt):
    weight_list.append(geo_mean_list[i]/geo_mean_sum)
```

```python
# 가중치 출력
print('가중치 (Weights)')
for i,j in zip(survey, weight_list):
    print('{} : {}'.format(i, j))
    #print('{} : {}'.format(i, round(j,3)))  # 소수점 3 자리수
```

```python
# 가중치 합 확인
su_ = sum(list(weight_list))
print(su_)
```

```python

survey = survey.append(pd.Series(weight_list, index=survey.columns),ignore_index=True)
survey #5행 - 기하평균, 6행 - 가중치
```

```python
# 데이터프레임을 csv로 저장
survey.to_csv('AHP_가중치산정_결과.csv',index=True , sep=',', encoding='cp949')
```
