```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

```python
cost = []
K = range(1, 5)
for num in K:
    kmode = KModes(n_clusters = num, n_init = 3)
    kmode.fit_predict(df)
    cost.append(kmode.cost_)

plt.plot(K, cost, 'bx-')
plt.xlabel('number of clusters')
plt.ylabel('cost')
plt.title('elbow method for finding optimal K')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('/content/for_folium.csv')
df.head()
```

```python
df = df.drop(labels='Unnamed: 0',axis=1)
df
```

```python
import requests
import json
```

```python
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
import folium

# params
center = [37.541, 126.986]
tiles = ['cartodbpositron', 'Stamen Toner', 'OpenStreetMap']

# visualization
m = folium.Map(
    location = [center[0], center[1]],
    zoom_start = 11,
    tiles = tiles[0]
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

# show
m
```

```python
m = folium.Map(
    location = [center[0], center[1]],
    zoom_start = 11,
    tiles = tiles[0]
)

m.choropleth(geo_data=seoul_geo,
             data=df,
             fill_color='Blues',
             color='gray',
             columns=('자치구','clusters'),
             fill_opacity=0.9,
             line_opacity=0.1,
             key_on='properties.name',
             legend_name="cluster"
            )
m
```
